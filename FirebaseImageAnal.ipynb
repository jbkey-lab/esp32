```python

col_names_Firebase = [
                    'computerImageFilePath'
                    'Comments' ,
                    'Field',
                    'FrameCountPerID',
                    'GreenLeaves'
                    'NumofTrackedObjectsPerImage',
                    'Range',
                    'Row',
                    'TotalFrameCount',
                    'dataCollectionInfo',
                    'dateString',
                    'id',
                    'mlNote',
                    'obdistance',
                    'obheight',
                    'objectID',
                    'oblabel',
                    'obwidth',
                    'pictureID',
                    'title',
                    'updatedDateString',
                    'urlPicture',
                    'printedPacket',
                    'printedPacketDate',
                    'printedPacketInital',
                    'printedPacketReprint'
                    ]

def download_image(folder_path, local_path):
    bucket = storage.bucket()
    blobs = bucket.list_blobs(prefix=folder_path)

    for blob in blobs:
        #print(f"File name is {blob}.json")
        blob.download_to_filename(local_path + blob.name.split('/')[-1]) 
        #print(f"Downloaded image to: {local_path}"+ blob.name.split('/')[-1])

def get_document_data_as_csv(user_id, filepath_to_save,local_image_path):
    db = firestore.client()
    doc_ref = db.collection("PAD").document(user_id).collection("Field")
    
    with open(filepath_to_save, mode='w', newline='', encoding='utf-8') as file:
        # Create a CSV writer object
        csv_writer = csv.writer(file)

        # Write the header row
        csv_writer.writerow(col_names_Firebase)

        # Stream the documents from the collection
        docs = doc_ref.stream()

        # Iterate over the documents
        for doc in docs:
            doc_dict = doc.to_dict()  # Convert document to dictionary
            
            # Check if 'pictureID' exists and is not empty
            if doc_dict.get('pictureID', '') != "":
                # Prepare row data
                row = [
                        local_image_path + doc_dict.get('id', '') + ".jpg",
                        doc_dict.get('Comments', ''),
                        doc_dict.get('Field', ''),
                        doc_dict.get('FrameCountPerID', ''),
                        doc_dict.get('GreenLeaves', ''),
                        doc_dict.get('NumofTrackedObjectsPerImage', ''),
                        doc_dict.get('Range', ''),
                        doc_dict.get('Row', ''),
                        doc_dict.get('TotalFrameCount', ''),
                        doc_dict.get('dataCollectionInfo', ''),
                        doc_dict.get('dateString', ''),
                        doc_dict.get('id', ''),
                        doc_dict.get('mlNote', ''),
                        doc_dict.get('obdistance', ''),
                        doc_dict.get('obheight', ''),
                        doc_dict.get('objectID', ''),
                        doc_dict.get('oblabel', ''),
                        doc_dict.get('obwidth', ''),
                        doc_dict.get('pictureID', ''),
                        doc_dict.get('title', ''),
                        doc_dict.get('updatedDateString', ''),
                        doc_dict.get('urlPicture', ''),
                        doc_dict.get('printedPacket', ''),
                        doc_dict.get('printedPacketDate', ''),
                        doc_dict.get('printedPacketInital', ''),
                        doc_dict.get('printedPacketReprint', '')
                ]  # Add other fields as needed
                # Write row to CSV
                
                csv_writer.writerow(row)
                #print(f"CSV Finished writing at {filepath_to_save}")
```
